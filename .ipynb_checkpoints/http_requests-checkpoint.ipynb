```python
import requests
from bs4 import BeautifulSoup # html parser

base_url = "https://www.reed.co.uk"
initial_url = "https://www.reed.co.uk/jobs/jobs-in-london?keywords=c%2B%2B&contract=True"

# response.status_code
# response.text
# print(dir(soup))

def get_number_of_pages(url):
    result_per_page = 20
    soup = BeautifulSoup(requests.get(url).text, "lxml")
    page_count = int(soup.find("span", class_="count").text.strip())
    count = page_count // result_per_page
    return count if page_count % result_per_page == 0 else count + 1
    
page_count = get_number_of_pages(initial_url)  
print(page_count)

all_job_urls = []
for page_no in range(1, page_count + 1):
    url = f"https://www.reed.co.uk/jobs/jobs-in-london?pageno={page_no}&keywords=c%2B%2B&contract=True"
    soup = BeautifulSoup(requests.get(url).text, "lxml")
    job_urls = [base_url + item.a["href"] for item in soup.find_all("h3", class_="title")]
    print(job_urls)
    all_job_urls.append(job_urls)

print(all_job_urls, len(all_job_urls))
# urls = [base_url + item.a["href"] for item in soup.find_all("h3", class_="title")]

# urls


# [i.a["href"] for i in soup.find_all("h3", class_="title")]

# urls = [i.a["href"] for i in soup.find_all("div", class_="job-title")]
# print(urls)
# for url in urls:
#     soup = BeautifulSoup(requests.get(url).text, "lxml")
#     print(soup.find('h1'))

# job = "https://www.totaljobs.com/job/python-developer/harnham-search-and-selection-ltd-job90997170"
# BeautifulSoup(requests.get(job).text, "lxml").find('h1')

# """
# <h3 

# </h3>
# """
```
